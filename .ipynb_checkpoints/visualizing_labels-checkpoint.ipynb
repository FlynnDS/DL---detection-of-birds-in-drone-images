```python
import json
import cv2
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
label_path = './scarecrow_dataset/train/annotations.json'

with open(label_path, 'r') as file:
    data = json.load(file)
```

```python
im_path = './scarecrow_dataset/train/images/DJI_0315.JPG'

image = cv2.imread(im_path)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
```

```python
img_annotation = [item for item in data if item['OriginalFileName'] == 'DJI_0315.JPG'][0]

for annotation in img_annotation['AnnotationData']:
    coords = annotation['Coordinates']
    
    x_min = int(min(pt['X'] for pt in coords))
    y_min = int(min(pt['Y'] for pt in coords))
    x_max = int(max(pt['X'] for pt in coords))
    y_max = int(max(pt['Y'] for pt in coords))

    cv2.rectangle(image_rgb, (x_min, y_min), (x_max, y_max), (64, 224, 208), 8)

plt.figure(figsize=(10, 8))
plt.imshow(image_rgb)
plt.show()
```
