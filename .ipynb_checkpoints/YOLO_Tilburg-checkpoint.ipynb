```python
!pip install ultralytics
import shutil
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt
```

```python
# datasets
folder_path = '/content/drive/MyDrive/JADS group assignments Semester 2/Deep Learning - Group Assignment/dev/data/'

# models
models = '/content/drive/MyDrive/JADS group assignments Semester 2/Deep Learning - Group Assignment/dev/models'

# logs
logs = '/content/drive/MyDrive/JADS group assignments Semester 2/Deep Learning - Group Assignment/dev/logs'

yolo_data = folder_path + 'data_yolo_format/data.yaml'
```

```python
model_stage_1 = YOLO("yolo11m.pt")
```

```python
model_stage_1.train(
    data=yolo_data,          # Path to your dataset configuration file

    # Hyperparameters
    epochs=100,                    # Number of epochs
    batch=4,                     # batch size
    imgsz=1280,                    # Image size
    freeze=10,                     # Freeze first 10 layers
    patience=10,
    nbs=64,
    #lr0= 0.01,
    #lrf=           0.1,
    #momentum=      0.937,
    #weight_decay=  0.0005,
    #warmup_epochs= 3.0,
    #warmup_momentum= 0.8,
    #warmup_bias_lr=  0.1,

    save=True,
    #deterministic=True,
    # Reproducability
    seed = 42
)
```

```python
import shutil

# Path to YOLOv8 trained model weights
trained_model_path = 'runs/detect/train/weights/last.pt' #update with the corresponding path

# Path to save in Google Drive with a custom name
custom_name = 'yolov11_freeze_10_modified_hp_32_epochs_last.pt'  # Replace with your desired name
destination_path = f'{models}/{custom_name}'

# Copy the best model to Google Drive
shutil.copy(trained_model_path, destination_path)

print(f'Model saved to {destination_path}')
```

```python
import shutil

# Path to YOLOv8 trained model weights
trained_model_path = 'runs/detect/train/weights/best.pt' #update with the corresponding path

# Path to save in Google Drive with a custom name
custom_name = 'yolov11_freeze_10_modified_hp_32_epochs_best.pt'  # Replace with your desired name
destination_path = f'{models}/{custom_name}'

# Copy the best model to Google Drive
shutil.copy(trained_model_path, destination_path)

print(f'Model saved to {destination_path}')
```

```python
metrics = model_stage_1.val(data=yolo_data, imgsz=1280, save=True, split='test')
print(metrics)
```

```python
# Load best or last checkpoint
model = YOLO("runs/detect/train/weights/best.pt")
# Evaluate on the test split defined in test.yaml
metrics = model.val(data=yolo_data, imgsz=1280, save=True, split='test')
print(metrics)
```
